```python
data1=pd.read_csv('/content/drive/MyDrive/heartrate_seconds_merged.csv')
```

```python
data1.head()
```

```python
data1.sample(12)
```

```python
data1.dtypes
```

```python
from datetime import datetime
#date_object = datetime.strptime(date_string, "%m/%d/%Y %I:%M:%S %p")
#data['ActivityDate'].astype('datetime64[ns]')
data1['Time'] = pd.to_datetime(data1['Time'], format='%m/%d/%Y %I:%M:%S %p')
```

```python
data1.dtypes
```

```python
data1.head()
```

```python
data1['date'] = pd.to_datetime(data1['Time']).dt.date
data1['time'] = pd.to_datetime(data1['Time']).dt.time
```

```python
data1.shape
```

```python
data1.duplicated().sum()
```

```python
data1.isna().sum()
```

```python
data1.head()
```

```python
data1['date']=pd.to_datetime(data1['date'])
```

```python
data1['Hours'] = pd.to_datetime(data1['time'], format='%H:%M:%S').dt.hour
```

```python
data1.head()
```

```python
data1.drop(columns=['time'], inplace=True)
```

```python
data1.shape
```

```python
# Group by 'date', 'Hours', and 'Id', then calculate the mean 'Value'
grouped_by_date_hours_id = data1.groupby(['date', 'Hours', 'Id'])['Value'].mean().reset_index()
grouped_by_date_hours_id.rename(columns={'Value': 'Avg_Value_By_Date_Hours_Id'}, inplace=True)

# Group by 'date' and 'Id', then calculate the mean 'Value'
grouped_by_date_id = data1.groupby(['date', 'Id'])['Value'].mean().reset_index()
grouped_by_date_id.rename(columns={'Value': 'Avg_Value_By_Date_Id'}, inplace=True)

# Merge the grouped data with the original data
merged_data = pd.merge(data1, grouped_by_date_hours_id, on=['date', 'Hours', 'Id'], how='left')
merged_data = pd.merge(merged_data, grouped_by_date_id, on=['date', 'Id'], how='left')

print("Merged Data:")
print(merged_data)
```

```python
merged_data.duplicated().sum()
```

```python
merged_data.to_csv('Final_heartrate_seconds_merged.csv', index=False)
```
